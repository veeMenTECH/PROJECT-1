# Correcting null cells by filling them with mean values
#### This is to ensure that data is not lost since it's always best to keep data rather than deleting it to prevent any form of misfortune from happening.
#### Hence, the null cells are filled with mean of all values in oder to establish a correlation with existing data in the cells

```python
everyColumn = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides','free sulfur dioxide', 'total sulfur dioxide','density', 'pH', 'sulphates', 'alcohol', 'quality']
```

```python
for column in everyColumn:
    data[column].fillna(data[column].mean(), inplace=True)
```
